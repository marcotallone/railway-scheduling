```python
from railway import *
import os
import ast

# Data directory
DATA = os.getcwd() + "/data/"

# List all files
files = os.listdir(DATA)
txtfiles = [f for f in files if f.endswith('.txt')]

for file in txtfiles:
    
    open_parameters = []
    with open(os.path.join(DATA, file), "r") as f:
        lines = f.readlines()
        counter = 0
        for line in lines:
            if counter > 0:
                open_parameters.append(ast.literal_eval(line))
            counter += 1

    print(f"Processing {file}...")
    T,N, N_coordinates, A, w_exp, w_inc, OD, phi,R, W_exp,J,J_a,p, E, C, t_int  = open_parameters

    # Constant values
    n = N[-1]
    periods = T
    jobs = J[-1]
    passengers = 2000
    K = 3

    # Create blueprint rail
    gen = Railway(n, periods, jobs, passengers, K)

    # Generate problem common parameters
    job_min_time, job_max_time = 1, 1
    job_min_length, job_max_length = 1, 1
    pause_min_time, pause_max_time = 0, 0
    min_demand, max_demand = 0.0, 1.0
    min_share, max_share = 1.0, 1.0
    min_capacity, max_capacity = 0.3, 0.7
    n_max_events = 0
    event_min_length, event_max_length = 1, 2
    gen.generate(
        job_min_time,
        job_max_time,
        job_min_length,
        job_max_length,
        pause_min_time,
        pause_max_time,
        min_demand,
        max_demand,
        min_share,
        max_share,
        min_capacity,
        max_capacity,
        n_max_events,
        event_min_length,
        event_max_length,
    )

    # Compute Aj
    Aj = {j: [a for a in A if j in J_a[a]] for j in J}

    # Convert E to dictionary
    E_dict = {}
    for key, value in E.items():
        E_dict[key] = dict(value)

    # Build tau dictionary
    tau = {}
    for key, value in J_a.items():
        tau[key] = t_int

    # Build null lambda dictionary
    null_lambda = {}
    for key, value in gen.Lambd.items():
        null_lambda[key] = 0.0
        
    # Ones beta
    ones_beta = {}
    for key, value in gen.beta.items():
        ones_beta[key] = 1.0
        
    # Coords list from dictionary
    list_coords = []
    for _, value in N_coordinates.items():
        list_coords.append(value)

    rail = Railway(
        stations=N[-1],
        periods=T,
        jobs=J[-1],
        passengers=2000,
        routes=3,
        C=C,
        Aj=Aj,
        R=R,
        pi=p,
        phi = phi,
        tau=tau,
        beta=ones_beta,
        Lambd=null_lambda,
        coords=list_coords
    )

    FILENAME = os.path.join(DATA, f"railway_N{n}_T{periods}_J{jobs}_P{passengers}_K{K}.json")
    rail.save(FILENAME)	
    print(f"✅ Dataset N={n}, T={periods}, J={jobs}, P={passengers}, K={K} created")
    del rail
    del gen
```
